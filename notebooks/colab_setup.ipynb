## 8. Backend ve Frontend'i BaÅŸlatma

````python
# Backend ve Frontend'i arka planda baÅŸlat
# Colab'da background process olarak Ã§alÄ±ÅŸtÄ±rÄ±yoruz

import os
import subprocess
import time

# Backend'i baÅŸlat
backend_process = subprocess.Popen(
    ['node', 'backend/server.js'],
    cwd='/content/bil482-project',
    stdout=open('/tmp/backend.log', 'w'),
    stderr=subprocess.STDOUT
)
print("ğŸ”„ Backend baÅŸlatÄ±ldÄ± (PID: {})".format(backend_process.pid))
time.sleep(5)

# Frontend'i baÅŸlat
frontend_process = subprocess.Popen(
    ['ng', 'serve', '--host', '0.0.0.0', '--port', '4200', '--disable-host-check'],
    cwd='/content/bil482-project/frontend',
    stdout=open('/tmp/frontend.log', 'w'),
    stderr=subprocess.STDOUT
)
print("ğŸ”„ Frontend baÅŸlatÄ±ldÄ± (PID: {}) - Angular build 2-3 dakika sÃ¼rebilir".format(frontend_process.pid))
time.sleep(20)  # Angular build iÃ§in biraz bekle

print("\nâœ… Backend ve Frontend baÅŸlatÄ±ldÄ±")
print("ğŸ“ Backend: http://localhost:3000")
print("ğŸ“ Frontend: http://localhost:4200")
print("\nğŸ“‹ LoglarÄ± gÃ¶rmek iÃ§in:")
print("   Backend:  !tail -20 /tmp/backend.log")
print("   Frontend: !tail -20 /tmp/frontend.log")
````
